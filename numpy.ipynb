Here’s a quick explanation of mean, mode, and median:

1. Mean (Average):
- Formula: (Sum of all values) ÷ (Number of values)
- Example: For [2, 4, 6], mean = (2+4+6)/3 = 4

2. Mode:
- The value that appears most frequently.
- Example: For [1, 2, 2, 3], mode = 2

3. Median:
- The middle value when data is sorted.
- If odd number of values: middle one.
- If even: average of two middle values.
- Example:  
  - [1, 3, 5] → median = 3  
  - [1, 3, 5, 7] → median = (3+5)/2 = 4
