Task 1: Create a Rectangle Class

Create a class Rectangle with methods to set length and width, and calculate area and perimeter.

Hint: Use instance variables self.length and self.width. Area = length × width, Perimeter = 2 × (length + width).



```python
class Rectangle:
    def __init__(self):
        self.length = 0
        self.width = 0

    def set_dimensions(self):
        self.length = float(input("Enter the length of the rectangle: "))
        self.width = float(input("Enter the width of the rectangle: "))

    def calculate_area(self):
        return self.length * self.width

    def calculate_perimeter(self):
        return 2 * (self.length + self.width)
```

```python
# Create object
rect = Rectangle()
```

```python
# Get user input
rect.set_dimensions()
```

```python
# Display results
print("\nRectangle Details:")
print("Length:", rect.length)
print("Width:", rect.width)
print("Area:", rect.calculate_area())
print("Perimeter:", rect.calculate_perimeter())
```

Task 2: Student Grade Manager

Create a class Student that stores a student’s name and three subject marks. Add methods to calculate the total marks and average marks.

Hint: Use instance variables for name and marks. Total = sum of all marks, Average = total / number of subjects.

```python
class Student:
    def __init__(self):
        self.name = ""
        self.mark1 = 0
        self.mark2 = 0
        self.mark3 = 0

    def get_details(self):
        self.name = input("Enter student name: ")
        self.mark1 = float(input("Enter marks for Subject 1: "))
        self.mark2 = float(input("Enter marks for Subject 2: "))
        self.mark3 = float(input("Enter marks for Subject 3: "))

    def calculate_total(self):
        return self.mark1 + self.mark2 + self.mark3

    def calculate_average(self):
        return self.calculate_total() / 3

    def show_details(self):
        print("\nStudent Details:")
        print("Name:", self.name)
        print("Subject 1 Marks:", self.mark1)
        print("Subject 2 Marks:", self.mark2)
        print("Subject 3 Marks:", self.mark3)
        print("Total Marks:", self.calculate_total())
        print("Average Marks:", round(self.calculate_average(), 2))
```

```python
# Create object
student1 = Student()
```

```python
# Get user input
student1.get_details()
```

```python
# Display results
student1.show_details()
```

Task 3: Temperature Converter

Create a class Temperature with methods to set temperature in Celsius and convert it to Fahrenheit and Kelvin.

Hint: Fahrenheit = (Celsius × 9/5) + 32, Kelvin = Celsius + 273.15.

```python
class Temperature:
    def __init__(self):
        self.celsius = 0.0

    def set_temperature(self):
        self.celsius = float(input("Enter temperature in Celsius: "))

    def to_fahrenheit(self):
        return (self.celsius * 9/5) + 32

    def to_kelvin(self):
        return self.celsius + 273.15

    def show_conversion(self):
        print("\nTemperature Conversion:")
        print("Celsius:", self.celsius, "°C")
        print("Fahrenheit:", round(self.to_fahrenheit(), 2), "°F")
        print("Kelvin:", round(self.to_kelvin(), 2), "K")
```

```python
# Create object
temp = Temperature()
```

```python

# Get user input
temp.set_temperature()
```

```python

# Display conversions
temp.show_conversion()
```

Task 5: Employee Management

Create a class Employee with methods to set employee details (name, ID, salary) and give a raise (increase salary by a given percentage).

Hint: Store name, ID, and salary as instance variables. For a 10% raise: new_salary = old_salary * 1.10.

```python
class Employee:
    def __init__(self):
        self.name = ""
        self.emp_id = ""
        self.salary = 0.0

    def set_details(self):
        self.name = input("Enter employee name: ")
        self.emp_id = input("Enter employee ID: ")
        self.salary = float(input("Enter employee salary: "))

    def give_raise(self, percentage):
        self.salary = self.salary * (1 + percentage / 100)

    def show_details(self):
        print("\nEmployee Details:")
        print("Name:", self.name)
        print("Employee ID:", self.emp_id)
        print("Current Salary:", round(self.salary, 2))


# Create object
emp = Employee()

# Get employee details
emp.set_details()

# Ask for raise percentage
raise_percent = float(input("Enter raise percentage: "))
emp.give_raise(raise_percent)

# Display updated details
emp.show_details()
```
